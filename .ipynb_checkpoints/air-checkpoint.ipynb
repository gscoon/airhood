### Database Stuff

```javascript
var connStr = 'postgres://postgres:ronnie@localhost/airbnb';

function insertAirRow(a){
    var q = "INSERT INTO air_listing (url, title, body) VALUES ($1, $2, $3)";
    var params = [a.url, a.title, a.body];
    pqQuery(q, params, function(err, results){})
}

function deleteAirRow(a, cb){
    var q = "DELETE FROM air_listing WHERE url = $1";
    var params = [a.url];
    pqQuery(q, params, cb)
}

function getAirRows(callback){
    var q = "SELECT * FROM air_listing";
    pqQuery(q, [], callback);
}

function updateAirROw(url, field, value){
    var q = "UPDATE air_listing SET " + field + " = ($1) WHERE url = ($2)";
    pqQuery(q, [value, url], function(err, res){
        if(err) console.log('db error: ', err)
    });
}

// craigslist

function insertCraigsRow(c, cb){
    var ts = moment().format("YYYY-MM-DD HH:mm:ss");
    var q = "INSERT INTO craigs_listing (url, details, timestamp) VALUES ($1, $2, $3)";
    var params = [c.url, c.details, ts];
    pqQuery(q, params, cb);
}

function getCraigsRows(callback){
    var q = "SELECT * FROM craigs_listing";
    pqQuery(q, [], callback);
}

function getCraigsURLs(url, callback){
    var q = "SELECT url FROM craigs_listing";
    pqQuery(q, [], callback);
}

function pqQuery(q, params, callback){
    pg.connect(connStr, function(err, client, done) {
        client.query(q, params, function(err, result){
            done();
            // if this is a select, only return the rows
            if(result && result.command == 'SELECT')
                result = result.rows;
            callback(err, result);
        });
    });
}
```

```javascript
var airArray = [];

function getAirDetails(a){
    // scrape airbnb listing page
    needle.get(a.url, function(err, res){
        if(err)
            return console.log('request error', err)
        
        a.body = res.body;
        a.location = '';
        
        var $ = cheerio.load(a.body);
        
        a.title = $('title').text();
        a.longitude = $('meta[property="airbedandbreakfast:location:longitude"]').attr('content');
        a.latitude = $('meta[property="airbedandbreakfast:location:latitude"]').attr('content');

        var aPattern = new RegExp(/eighborhood&quot;:&quot;([a-zA-Z _-]+)/);
        aMatches = a.body.match(aPattern);
        if(aMatches != null)
            a.location = aMatches[1];
        
        airArray.push(a);
        
        // first delete if it exists, then add it!
        deleteAirRow(a, function(err, res){
            insertAirRow(a, function(err, res){}); // to database
        });
        
        
        console.log('pulled: ' + a.url);
    });
}
```

```javascript
var p = {};

function openPage(url){
    var page = p[url]['page'];
    page.open(url, function(status){
        console.log('status: ', status, url);
        if(status)
            setTimeout(function(){
                page.evaluate(getDocument, handleDocument);
            }, 1000); // wait a second before
    });
}

function getDocument(){
    var o = {body: document.documentElement.outerHTML, title: document.title, url: window.location.href.split("?")[0]};
    return o;
}

function handleDocument(r){
    var ph = p[r.url]['ph'];
    console.log('handle url: ' + r.url, moment().format('YYYY-MM-DD H:mm:ss'));
    updateAirROw(r.url, 'body', r.body);
    updateAirROw(r.url, 'title', r.title);
    ph.exit();
}

function scrapeAirPages(){
    getAirRows(function(err, rows){
    var waitIndex = 0;
        rows.forEach(function(row, i){
            // if (i>1) return true;
            setTimeout(function(){
                phantom.create(function(ph){
                    ph.createPage(function (page) {
                        p[row.url] = {
                            page: page,
                            ph: ph
                        };
                        openPage(row.url); 
                    });
                });
            }, waitIndex * 4000); // wait 4 secs
            waitIndex++;
        
        })
    })
}

var n = null;
```

```javascript
function makeid(len){
    var text = "";
    var possible = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789";
    for( var i=0; i < len; i++ )
        text += possible.charAt(Math.floor(Math.random() * possible.length));
    return text;
}
```

```javascript
var mag = {long: -71.1113697, lat: 42.360491};
var maxD = 0.028985507; // 2 miles in lat-long

airData = [];
airProx = [];

getAirRows(function(err, rows){
    airData = rows;
    rows.forEach(function(row, i){
        var A = mag.long - row.details.longitude;
        var B = mag.lat - row.details.latitude;
        var C = Math.pow(Math.pow(A,2) + Math.pow(B,2), .5); //pythag
        row.details.distance = C * 69; // miles
        updateAirROw(row.url, 'details', row.details);
        if(C <= maxD) airProx.push(i)
    });
});
```

```javascript
var keys = [];
var vals = [];

var airHood = {};
airProx.forEach(function(key){
    var hood = airData[key].details.hood;
    if(typeof airHood[hood] == 'undefined') airHood[hood] = 0;
    airHood[hood]++;
});

for(var k in airHood){
    keys.push(k);
    vals.push(airHood[k]);
};

var plotlyToken = 'bkgynz91em';
var plt1 = {x: keys, y: vals, type: "bar"};

var figure = { 'data': [plt1] };

var imgOpts = {format: 'png'};

plotly.getImage(figure, imgOpts, function (err, imageStream) {
    console.log(err);
    var imgstring = '';
    var chunks = [];
    
    var rand = makeid(3);
    console.log('Random: ' + rand);
    
    imageStream.on('data', function(chunk) {
        chunks.push(chunk);
    });
    
    imageStream.on('end', function(){
        var result = Buffer.concat(chunks);
        console.log('Image set: ' + rand);
        lastImage = result.toString('base64');
    });
});
```

### Craigslist scraping

```javascript
(function(){
    var base = 'https://boston.craigslist.org';
    var frontPageURL = 'https://boston.craigslist.org/search/gbs/nfa?max_price=2000&query=private+entrance&hasPic=1&min_price=1000';
    needle.get(frontPageURL, handleFrontPage);
    
    
    function handleFrontPage(err, res){
        var b = res.body;
        var $ = cheerio.load(b);
        var As = $('.row a.i');
        var waitIndex = 0;
        // loop through each anchor
        As.each(function(i, v){
            //if(i > 1) return true;
            setTimeout(function(){
                var url = $(v).attr('href');
                if(url.indexOf(':') == -1)  url = base + url;
                console.log(url, moment().format("YYYY-MM-DD HH:mm:ss"));
                // get scrape anchor page
                needle.get(url, function(err, res){
                    var $ = cheerio.load(res.body);
                    var c = {url:url};

                    var d = {
                        latitude: parseFloat($('#map').attr('data-latitude')),
                        longitude: parseFloat($('#map').attr('data-longitude')),
                        title: $('title').text(),
                        price: $('.price').text(),
                        inside: $('.attrgroup').eq(0).find('span').eq(0).text(),
                    }

                    d.distance = calcDistance(d.latitude, d.longitude);
                    c.details = d;

                    insertCraigsRow(c, function(err, res){
                        // if(err) console.log(err);
                    });
                });
            }, waitIndex * 1000); // wait 2 seconds between
            
            waitIndex++;
        });

    }
    

    function calcDistance(lat, long){
        var A = magazine.long - long;
        var B = magazine.lat - lat;
        var C = Math.pow(Math.pow(A,2) + Math.pow(B,2), .5); //pythag
        return C * 69; // miles
    }
    
})()
```
